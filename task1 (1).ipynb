```python
from bs4 import BeautifulSoup 
import requests

def requestSeatNumber(seatNumber):
    response = requests.get(f'https://shbabbek.com/natega/{seatNumber}')
    return response

def check_seatNumber(seatNumber):
    response = requestSeatNumber(seatNumber)
    soup = BeautifulSoup(response.content)
    #sent = soup.find('p',attrs={'style':"font-size: 14px;color: red; margin-top: 11px;"}).text==''
    #print(sent)
    sent = soup.find('p',attrs={'style':"font-size: 14px;color: red; margin-top: 11px;"})

    if sent is not None:
        if sent.text=='':
            #getStudentInfo()
            student_inform(soup)
        else: return seatNumber
    else: return 0

def student_inform(soup):
    info = soup.find_all('tr', class_='dataTable__title')
    subject_tags = info[1].parent.parent.select('tbody > tr')
    personal_info = info[0].parent.parent.find('tbody').find_all('tr')
    seatNumber = int(personal_info[0].find_all('td')[1].get_text().strip())
    name = list(personal_info[1].children)[3].text.strip()
    school = personal_info[2].find_all('td')[1].get_text().strip()
    schoolSection  = personal_info[3].find_all('td')[1].get_text().strip()
    department  = personal_info[4].find_all('td')[1].get_text().strip()
    total = float(personal_info[5].find_all('td')[1].get_text().strip())
    percentage = float(personal_info[6].find_all('td')[1].strong.text.strip()[:-1]) # to remove %
    trail_1_2 = personal_info[8].find_all('td')[1].get_text().strip()
    Arabic_grade = subject_tags[0].find_all('td')[1].get_text().strip()
    
    English_grade = subject_tags[1].find_all('td')[1].get_text().strip()
    scd_lang_grade = subject_tags[2].find_all('td')[1].get_text().strip()
    pure_math_grad = subject_tags[3].find_all('td')[1].get_text().strip()
    history_grade = subject_tags[4].find_all('td')[1].get_text().strip()
    Geographic_grade = subject_tags[5].find_all('td')[1].get_text().strip()
    philoSo_grade = subject_tags[6].find_all('td')[1].get_text().strip()
    sicolo_grade = subject_tags[7].find_all('td')[1].get_text().strip()
    chemistry_grade = subject_tags[8].find_all('td')[1].get_text().strip()
    bio_grade = subject_tags[9].find_all('td')[1].get_text().strip()
    Geo_grade = subject_tags[10].find_all('td')[1].get_text().strip()
    Applied_math_grade = subject_tags[11].find_all('td')[1].get_text().strip()
    pysics_grade = subject_tags[12].find_all('td')[1].get_text().strip()
    
    relg = subject_tags[14].find_all('td')[1].get_text().strip()
    citizen_education = subject_tags[15].find_all('td')[1].get_text().strip()
    econm = subject_tags[16].find_all('td')[1].get_text().strip()
    
    with open('StudentGradesFirstTrial.csv', 'a', encoding='utf-8-sig') as f2:
        f2.write(f"{seatNumber},{name},{school},{schoolSection},{department},{total},{percentage},{trail_1_2},{Arabic_grade[0]},{English_grade},{scd_lang_grade},{pure_math_grad},{history_grade},{Geographic_grade},{philoSo_grade},{sicolo_grade},{chemistry_grade},{bio_grade},{Geo_grade},{Applied_math_grade},{pysics_grade},{relg},{citizen_education},{econm}\n")
        
if __name__ == '__main__':
    print("Thanawya 3ama Results from Shbabbek.com Website")
    with open('StudentGradesFirstTrial.csv', 'w', encoding='utf-8-sig') as f:
        f.write('seat_number,name,school,educational_administration,branch,total,percentage,status,arabic,first_foreign_lang,second_foreign_lang,pure_math,history,geography,philosophy,psychology,chemistry,biology,geology,applied_math,physics,religion,citizenship_education,economics\n')
    start = int(input("Enter the start of the range: "))
    finish = int(input("Enter the end of the range: ")) + 1
    for num in range(start, finish):
        print(f"Checking {num}...")
        num_status = check_seatNumber(num)
        if not num_status: print(f'{num} is valid')
        else: print(f'{num} is invalid')
    input("\nEnter any Character to Exit\n")
```
