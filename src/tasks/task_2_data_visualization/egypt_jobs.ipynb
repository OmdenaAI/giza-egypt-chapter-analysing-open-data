```python
df = data.copy()
df.shape
```

```python
df.dtypes.value_counts()
```

```python
plt.figure(figsize=[15,10])
first_20_jobs = df.Title.value_counts().head(20).index
count = df.Title.value_counts().head(20)

plt.barh(first_20_jobs, count)

plt.title('Top 20 Jobs Needed in Egypt')
plt.xticks(fontsize= 11)
plt.yticks(fontsize= 12)
plt.xlabel(' Requested time /Job')
```

```python
plt.figure(figsize=[15,10])
top_companies = df.Company.value_counts().head(20).index
count = df.Company.value_counts().head(20)

plt.barh(top_companies, count)

plt.title('Top 20 companies Egypt 2020', fontsize= 15)
plt.xticks(fontsize= 11)
plt.yticks(fontsize= 12)
plt.xlabel(' Requested times / company', fontsize=12);
```

```python
plt.figure(figsize=[15,10])
Years_Exp = df.YearsExp.value_counts().head(20).index
count = df.YearsExp.value_counts().head(20)

plt.barh(Years_Exp, count)

plt.title('Experience', fontsize= 15)
plt.xticks(fontsize= 11)
plt.yticks(fontsize= 12)
plt.xlabel('number of jobs needed such exeperience level', fontsize=12);
```
