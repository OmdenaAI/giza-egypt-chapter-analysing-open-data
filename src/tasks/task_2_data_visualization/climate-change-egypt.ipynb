# Climate Change Indicators - Egypt

```python
#importing libraries
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")

import plotly.express as px
```

```python
df = pd.read_csv('climate-change_egy.csv')
df = df.iloc[1: , :]
df
```

```python

df['Value'] = df['Value'].astype(float)
df['Year'] = df['Year'].astype(int)
df
```

### Land

```python
temp = df[df['Indicator Name'] == 'Agricultural land (sq. km)']
temp = temp.sort_values(by=['Year'], )
temp = temp.reset_index(drop=True)


# Agriculture land (sq. km)


#sns.lineplot(data=temp, x='Year', y='Value').set(title='Agricultural land( sq. km)')
p= sns.relplot(data=temp, x='Year', y='Value', kind='line', height=5, aspect=2)
p.set(title='Agricultural land (sq. km)')

fig = px.line(temp, x="Year", y="Value", title='Agricultural land (sq. km)' , width=900, height=500)

fig.show()
```

```python
temp = df[df['Indicator Name'] == 'Arable land (% of land area)']
temp = temp.sort_values(by=['Year'], )
temp = temp.reset_index(drop=True)

temp

# Agriculture land (sq. km)


p= sns.relplot(data=temp, x='Year', y='Value', kind='line', height=5, aspect=2)
p.set(title='Arable land (% of land area)')

fig = px.line(temp, x="Year", y="Value", title='Arable land (% of land area)' , width=900, height=500)

fig.show()
```

```python
temp = df[df['Indicator Name'] == 'Forest area (sq. km)']
temp = temp.sort_values(by=['Year'], )
temp = temp.reset_index(drop=True)

temp

# Forest area (sq. km))
plt.figure(figsize=(25,8))
sns.barplot(data=temp, x='Year', y='Value',)

fig = px.bar(temp, x="Year", y="Value", title='Forest area (sq. km)' , width=900, height=500)

fig.show()
```

#### Summary

- Agricultural land: 25.68k (1961) | 38.3k (2018)

- Arable land(% of land area):  2.59% (1961) | 2.92% (2018)

- Forest area:  438 sq. km (1990) | 449.8 sq. km (2020) | Gradually increased from 1990 to 2010 (656 sq. km) then decreased to 449.88 sq. km
