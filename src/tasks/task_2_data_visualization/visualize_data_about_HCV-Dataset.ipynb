The higher the staging is, the bigger is the advancement of the desease. Indeed,  The fibrosis status could be recognized as a key predictor of the adverse of  hepatic outcomes and mortality.
=> In untreated HCV-infected patients, liver fibrosis may progress over time.

## Relations Blood / Variables

In this part, let's first divise the Datasert into 4 groups based on the histologic grades : from 1 to 4

```python
minimal_advancement_group= data[data['Baselinehistological staging'] == 1]
mild_advancement_group= data[data['Baselinehistological staging'] == 2]
moderate_advancement_group= data[data['Baselinehistological staging'] == 3]
severe_advancement_group= data[data['Baselinehistological staging'] == 4]
```

```python
minim_cout=minimal_advancement_group.count()
mild_cout=mild_advancement_group.count()
moderate_cout=moderate_advancement_group.count()
severe_cout=severe_advancement_group.count()
```
