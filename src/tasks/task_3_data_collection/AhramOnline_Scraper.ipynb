# Import needed libraries

```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

# Get all the page Content

```python
src = requests.get("https://english.ahram.org.eg/UI/Front/Search.aspx?Text=accidents&PID=1&StartRowIndex=140").content
soup= BeautifulSoup(src , "lxml")
```

# Select some specific needed tags

```python
article_src=[]
articles_src =soup.find_all("div",{"class":"num_result"})

article_title=[]
articles_title= soup.find_all("h5")

article_date=[]
articles_date= soup.find_all("div",{"class":"name_result"})

links=[]
```

# Extract only text from the scaped data

```python
for i in range(len(articles_src)):
    article_src.append(articles_src[i].text.replace("\n",""))
    article_title.append(articles_title[i].text)
    article_date.append(articles_date[i].text.replace("\n",""))
    links.append(articles_title[i].find("a").attrs['href'].replace("../../","https://english.ahram.org.eg/"))

# Iterate over links to get the atricle content
Article_content=[]
content=""
for link in links:
    src = requests.get(link).content
    soup = BeautifulSoup(src,"lxml")
    Articles_content = soup.find("div",{"class":"col-12 nopadding mar-top fixcontent","id":"ContentPlaceHolder1_divContent"})
    if Articles_content is not None:
        Articles_content=Articles_content.text.replace("\n","")
    else:
        Articles_content= "None"
    
    Article_content.append(Articles_content)    
    
    #print(Articles_content)
    Article_content
```

```python
Articles_Scraper = pd.DataFrame(
    {'Article Date': article_date,
     'Atricle Source': article_src,
     'Article Title': article_title,
     'Article Link' : links , 
     'Article content': Article_content
    })

#Articles_Scraper['Article content'].str.strip()
Articles_Scraper
```

```python
Articles_Scraper.to_csv('ACCident_Articles140.csv')
```
