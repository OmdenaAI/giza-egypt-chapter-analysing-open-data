# Import needed libraries

```python
import requests
import pandas as pd
import matplotlib.pyplot as  plt
import datetime
```

# The main fuction of request, respond, and extracting data

```python
def get_yearly_rates(amount , currency , converted_curr , amount_of_days):
    #setting the start date
    today_date = datetime.datetime.now()
    date_1year = today_date - datetime.timedelta(days = 1 * amount_of_days)
    
    #Creating a data request
    url = f'http://api.exchangerate.host/timeseries'
    payload={'base':currency , 'amount':amount , 'start_date':date_1year.date() ,'end_date':today_date.date() }
    response = requests.get(url , params=payload)
    data = response.json()
    
    # creating an empty dict to append new data to it 
    stored_Currency={}
    history_rate = []
    
    # Extracting date in a fitted shape
    for item in data['rates']:
        currency_rate = data['rates'][item][converted_curr]
        stored_Currency[item] =[currency_rate]
        history_rate.append(currency_rate)
        
    #convert to dataframe
    df = pd.DataFrame(stored_Currency).transpose()
    df.columns = ['Daily_Rate']
    pd.set_option('display.max_rows' , None)
    
    
    #Visualizing the daily currency change rate  
    plt.figure(figsize=(15,5))
    #plt.plot(Finished_dataset['Daily_Rate'])
    plt.plot(history_rate)
    plt.xlabel('Days')
    plt.show()
    
    return df
    
```

```python
dataset_of2019 = get_yearly_rates(1 , 'USD' , 'EGP' , 819)
dataset_of2019
print(dataset_of2019)
dataset_of2019.to_csv('dataset_of2019.csv')
```

```python
dataset_of2020 = get_yearly_rates(1 , 'USD' , 'EGP' , 731)
dataset_of2020
print(dataset_of2020)
dataset_of2020.to_csv('dataset_of2020.csv')
```

```python
dataset_of2021 = get_yearly_rates(1 , 'USD' , 'EGP' , 365)
dataset_of2021
dataset_of2021.to_csv('dataset_of2021.csv')
```
